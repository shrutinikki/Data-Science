```python
#q1
cars_data=pd.read_csv('autos_small.csv')
cars_data.head()
```

```python
#q2
cars_data.isnull().any()
```

```python
cars_data.fillna(method='bfill',inplace=True)
```

```python
cars_data.isnull().any()
```

```python
#q3
cars_data['monthOfRegistration'].value_counts()
```

```python
cars_data['monthOfRegistration']=cars_data['monthOfRegistration'].replace(0,cars_data['monthOfRegistration'].median())
```

```python
cars_data['monthOfRegistration'].value_counts()
```

```python
#converting catagories data to dummy data
cat_cols = ['name','model','gearbox','fuelType','brand','vehicleType']
cars_data = pd.get_dummies(cars_data, columns = cat_cols, drop_first=True)
```

```python
#checking for feature selection
y = cars_data.iloc[:, 0].values
X = cars_data.iloc[:, 1:].values

tree_clf = ExtraTreesClassifier()
tree_clf.fit(X, y)

tree_clf.feature_importances_
tree_clf.classes_
```

```python
importances = tree_clf.feature_importances_
feature_names = cars_data.iloc[:, 1:].columns.tolist()
feature_imp_dir = dict(zip(feature_names, importances))
features = sorted(feature_imp_dir.items(), key=lambda x: x[1], reverse=True)
features
```

```python
model = SelectFromModel(tree_clf, prefit = True)
x_update = model.transform(X)
print('Total Features count:', np.array(X).shape[1])
print('Selected Features: ' ,np.array(x_update).shape[1])
```

```python
y = cars_data.iloc[:, 0].values

#split dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(x_update, y, test_size = 0.2, random_state = 0)

#Standardisation
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#q4

#KNN
knn = KNeighborsRegressor(n_neighbors=10, weights='distance')
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

sqrt(mean_squared_error(y_test, y_pred))
```

```python
#Multiple Linear Regression
multi_lr_model = LinearRegression()

# fit the model with X and y

multi_lr_model.fit(X_train, y_train)

X_test = np.array(X_test)

# get predictions
y_pred = multi_lr_model.predict(X_test)
```

```python
rmse = sqrt(mean_squared_error(y_test, y_pred))

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

adj_r2 = 1 - float(len(y)-1)/(len(y)-len(multi_lr_model.coef_)-1)*(1 - r2)

rmse, r2, adj_r2, multi_lr_model, multi_lr_model.coef_, multi_lr_model.intercept_
```
