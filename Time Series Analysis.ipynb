```python
import pandas as pd
import numpy as np
from nsepy import get_history
import datetime as dt

start = dt.datetime(2015, 1, 1)

end = dt.datetime.today()

hdfc = get_history(symbol='HDFC', start = start, end = end) #retrieving the data
hdfc.index = pd.to_datetime(hdfc.index)
hdfc.head()
```

#### 1. In hdfc, drop all the columns except Open, High, Low and Close. Call the new df hdfc_ohlc

```python
hdfc.drop(['Prev Close','Symbol','Series','Last','VWAP','Volume','Turnover','Deliverable Volume','%Deliverble','Trades'],axis=1,inplace=True)
hdfc
```

```python
hdfc_ohlc = hdfc[['Open','High','Low','Close']]
hdfc_ohlc.head()
```

#### 2. Use the dataframe hdfc_ohlc created in qn 1 to create hdfc_mean with mean prices every year. 

```python
hdfc_mean=hdfc_ohlc.groupby(hdfc_ohlc.index.year).mean()
hdfc_mean
```

#### 3. Group the hdfc_ohlc to get the mean prices per week per year. The resulting dataframe hdfc_weekly should have mean price data for week 1 to 52 for years 2015, 2016 and 2017. 

Hint: You may have to groupby open_prices.index.week and open_prices.index.year to ensure that you get data per week per year.

```python
hdfc_weekly=hdfc_ohlc.groupby([hdfc_ohlc.index.year,hdfc_ohlc.index.week]).mean()
hdfc_weekly
```

#### 4. Create a dataframe infy and find total volumn by month, save it in infy_volume. Infy_volume should be a multi-index series with tow levels of index - Year and Month.

```python
infy_data = get_history(symbol='INFY', start = start, end = end)
infy_data.index = pd.to_datetime(infy_data.index)
infy_data.insert(0,'year',infy_data.index.year)
infy_data.insert(0,'month',infy_data.index.month)
infy_volume=infy_data.groupby([infy_data.year,infy_data.month]).Volume.sum()
infy_volume.head(10)
```

#### 5. Create a series infy_pctchange from the infy_volume dataframe created in qn 4. It should have two levels of index, Year and Month and the value should be % change in volume compared to the previous row.

```python
infy_pctchange=infy_volume

i=0
while i<len(infy_volume)-1:
    pct=((infy_volume[i+1]-infy_volume[i])/infy_volume[i]) *100
    infy_pctchange[i]=pct
    i=i+1
infy_pctchange
```

```python
infy_pctchange = infy_volume.pct_change()
```

#### 7. Create dataframe wipro and keep columns Open, High, Low, Close. Now find minimum close price each month by year.

```python
wipro = get_history(symbol='WIPRO', start = start, end = end) #retrieving the data
wipro.index = pd.to_datetime(wipro.index)
wipro.head()
```

```python
wipro['Close'].groupby([wipro.index.year,wipro.index.month]).min()
```
