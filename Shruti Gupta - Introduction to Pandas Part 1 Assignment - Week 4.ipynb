### 8. Find out in which stadium the maximum number of matches have been played and in which stadium minimum? 
### What is the highest and the lowest number? Show the code.


```python
maxmatches = 0
minmatches = 0
maxven = matches['venue'].max() 
minven = matches['venue'].min()
print("The stadium that has had maximum matches played is ",maxven," and number of matched played is ",np.where(matches['venue']==maxven,(maxmatches+1),maxmatches).sum())
print("The stadium that has had minimum matches played is ",minven," and number of matched played is ",np.where(matches['venue']==minven,(maxmatches+1),maxmatches).sum())
```
