### 9. Pivot the following dataframe to create df_pivoted with four columns: category, A, B, C. Aggregate the values on sum.
df = pd.DataFrame({'Name' : ['one', 'one', 'two', 'three'] * 3,
                   'Category' : ['A', 'B', 'C'] * 4,
               'Value' : range(12)})

```python
df = pd.DataFrame({'Name' : ['one', 'one', 'two', 'three'] * 3,
                   'Category' : ['A', 'B', 'C'] * 4,
               'Value' : range(12)})
df
```

```python
df_pivoted = pd.pivot_table(df, index='Name', columns='Category', values = 'Value', aggfunc='sum')
df_pivoted
```

### 10. Use the df_pivoted generated in qn 9 and add three columns A_pct, B_pct, C_pct to represent the percentage values of each cell per column. Hint: use df.apply(lambda x: .....)

```python
df_pivoted['A_pct']=df_pivoted['A'].apply(lambda x: (x/(df_pivoted['A'].sum()))*100)
df_pivoted['B_pct']=df_pivoted['B'].apply(lambda x: (x/(df_pivoted['B'].sum()))*100)
df_pivoted['C_pct']=df_pivoted['C'].apply(lambda x: (x/(df_pivoted['C'].sum()))*100)
```
